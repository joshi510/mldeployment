```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

# Load dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\Salary_data.csv")

X = df[["YearsExperience"]]   # feature must be 2D
y = df["Salary"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Save model
joblib.dump(model, "salary_model.joblib")

print("Model trained & saved ✅")
```
